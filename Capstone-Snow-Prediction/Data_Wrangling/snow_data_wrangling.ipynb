Data Cleaning
 ■ NA or missingvalues
 ■ Duplicates

```python
#Forward filling NA values for soil moisture and temperature columns
snow_data.loc[:, ['soilmoisture_min', 'soilmoisture_max', 'soilmoisture_avg']] = snow_data.loc[:, ['soilmoisture_min', 'soilmoisture_max', 'soilmoisture_avg']].ffill()
snow_data.loc[:, ['soiltemp_min', 'soiltemp_max', 'soiltemp_avg']] = snow_data.loc[:, ['soiltemp_min', 'soiltemp_max', 'soiltemp_avg']].ffill()
```

```python
snow_data.info()
```

```python
#Prepare date data for time-window features
snow_data['date'] = pd.to_datetime(snow_data['date'])
snow_data.set_index('date', inplace=True)
```

```python
#Calculating the rolling statistics 
#Avgerages:  7 day rolling  
snow_data['7d_airtemp_avg'] = snow_data['airtemp_obs'].rolling(window=7).mean()
snow_data['7d_precip_avg'] = snow_data['precip_accumulation'].rolling(window=7).mean()
snow_data['7d_snowdepth_avg'] = snow_data['snowdepth'].rolling(window=7).mean()
snow_data['7d_soiltemp_avg'] = snow_data['soiltemp_obs'].rolling(window=7).mean()

# Average:  30-day rolling 
snow_data['30d_airtemp_avg'] = snow_data['airtemp_obs'].rolling(window=30).mean()
snow_data['30d_precip_avg'] = snow_data['precip_accumulation'].rolling(window=30).mean()
snow_data['30d_snowdepth_avg'] = snow_data['snowdepth'].rolling(window=30).mean()
snow_data['30d_soiltemp_avg'] = snow_data['soiltemp_obs'].rolling(window=30).mean()

#Standard Deviation:  7 day rolling  
snow_data['7d_airtemp_std'] = snow_data['airtemp_obs'].rolling(window=7).std()
snow_data['7d_precip_std'] = snow_data['precip_accumulation'].rolling(window=7).std()
snow_data['7d_snowdepth_std'] = snow_data['snowdepth'].rolling(window=7).std()
snow_data['7d_soiltemp_std'] = snow_data['soiltemp_obs'].rolling(window=7).std()

# Standard Deviation:  30-day rolling 
snow_data['30d_airtemp_std'] = snow_data['airtemp_obs'].rolling(window=30).std()
snow_data['30d_precip_std'] = snow_data['precip_accumulation'].rolling(window=30).std()
snow_data['30d_snowdepth_std'] = snow_data['snowdepth'].rolling(window=30).std()
snow_data['30d_soiltemp_std'] = snow_data['soiltemp_obs'].rolling(window=30).std()

#Variance:  7 day rolling  
snow_data['7d_airtemp_var'] = snow_data['airtemp_obs'].rolling(window=7).var()
snow_data['7d_precip_var'] = snow_data['precip_accumulation'].rolling(window=7).var()
snow_data['7d_snowdepth_var'] = snow_data['snowdepth'].rolling(window=7).var()
snow_data['7d_soiltemp_var'] = snow_data['soiltemp_obs'].rolling(window=7).var()

# Variance:  30-day rolling 
snow_data['30d_airtemp_var'] = snow_data['airtemp_obs'].rolling(window=30).var()
snow_data['30d_precip_var'] = snow_data['precip_accumulation'].rolling(window=30).var()
snow_data['30d_snowdepth_var'] = snow_data['snowdepth'].rolling(window=30).var()
snow_data['30d_soiltemp_var'] = snow_data['soiltemp_obs'].rolling(window=30).var()


#Sum:  7 day rolling  
snow_data['7d_airtemp_sum'] = snow_data['airtemp_obs'].rolling(window=7).sum()
snow_data['7d_precip_sum'] = snow_data['precip_accumulation'].rolling(window=7).sum()
snow_data['7d_snowdepth_sum'] = snow_data['snowdepth'].rolling(window=7).sum()
snow_data['7d_soiltemp_sum'] = snow_data['soiltemp_obs'].rolling(window=7).sum()

# Sum:  30-day rolling 
snow_data['30d_airtemp_sum'] = snow_data['airtemp_obs'].rolling(window=30).sum()
snow_data['30d_precip_sum'] = snow_data['precip_accumulation'].rolling(window=30).sum()
snow_data['30d_snowdepth_sum'] = snow_data['snowdepth'].rolling(window=30).sum()
snow_data['30d_soiltemp_sum'] = snow_data['soiltemp_obs'].rolling(window=30).sum()


#Median:  7 day rolling  
snow_data['7d_airtemp_median'] = snow_data['airtemp_obs'].rolling(window=7).median()
snow_data['7d_precip_median'] = snow_data['precip_accumulation'].rolling(window=7).median()
snow_data['7d_snowdepth_median'] = snow_data['snowdepth'].rolling(window=7).median()
snow_data['7d_soiltemp_median'] = snow_data['soiltemp_obs'].rolling(window=7).median()

# Median:  30-day rolling 
snow_data['30d_airtemp_median'] = snow_data['airtemp_obs'].rolling(window=30).median()
snow_data['30d_precip_median'] = snow_data['precip_accumulation'].rolling(window=30).median()
snow_data['30d_snowdepth_median'] = snow_data['snowdepth'].rolling(window=30).median()
snow_data['30d_soiltemp_median'] = snow_data['soiltemp_obs'].rolling(window=30).median()


#min:  7 day rolling  
snow_data['7d_airtemp_min'] = snow_data['airtemp_obs'].rolling(window=7).min()
snow_data['7d_precip_min'] = snow_data['precip_accumulation'].rolling(window=7).min()
snow_data['7d_snowdepth_min'] = snow_data['snowdepth'].rolling(window=7).min()
snow_data['7d_soiltemp_min'] = snow_data['soiltemp_obs'].rolling(window=7).min()

# min:  30-day rolling 
snow_data['30d_airtemp_min'] = snow_data['airtemp_obs'].rolling(window=30).min()
snow_data['30d_precip_min'] = snow_data['precip_accumulation'].rolling(window=30).min()
snow_data['30d_snowdepth_min'] = snow_data['snowdepth'].rolling(window=30).min()
snow_data['30d_soiltemp_min'] = snow_data['soiltemp_obs'].rolling(window=30).min()


#max:  7 day rolling  
snow_data['7d_airtemp_max'] = snow_data['airtemp_obs'].rolling(window=7).max()
snow_data['7d_precip_max'] = snow_data['precip_accumulation'].rolling(window=7).max()
snow_data['7d_snowdepth_max'] = snow_data['snowdepth'].rolling(window=7).max()
snow_data['7d_soiltemp_max'] = snow_data['soiltemp_obs'].rolling(window=7).max()

# max:  30-day rolling 
snow_data['30d_airtemp_max'] = snow_data['airtemp_obs'].rolling(window=30).max()
snow_data['30d_precip_max'] = snow_data['precip_accumulation'].rolling(window=30).max()
snow_data['30d_snowdepth_max'] = snow_data['snowdepth'].rolling(window=30).max()
snow_data['30d_soiltemp_max'] = snow_data['soiltemp_obs'].rolling(window=30).max()
```

```python
snow_data.fillna(method='bfill', inplace=True)
snow_data.reset_index(inplace = True)
print(snow_data)
```
