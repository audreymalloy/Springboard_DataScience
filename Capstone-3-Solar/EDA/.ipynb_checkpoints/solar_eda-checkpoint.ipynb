## 2. Identify Relationships and Trends  

### Correlation Analysis

```python
p1_gen.dtypes
p1_weather.dtypes
```

```python
# Correlation heatmap for Plant 1 Generation

correlation_features = p1_gen[['AC_POWER','DC_POWER','DAILY_YIELD','TOTAL_YIELD']]
correlation_matrix = correlation_features.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Plant 1 Generation Data')
plt.show()

correlation_features_weather = p1_weather.select_dtypes(include=['float64', 'int64'])
correlation_matrix_weather = correlation_features_weather.corr()

# Correlation heatmap for Plant 1 Weather
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_weather.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Plant 1 Weather')
plt.show()
```

```python

correlation_features = p2_gen[['AC_POWER','DC_POWER','DAILY_YIELD','TOTAL_YIELD']]
correlation_matrix = correlation_features.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Plant 2 Generation Data')
plt.show()

correlation_features_weather = p2_weather.select_dtypes(include=['float64', 'int64'])
correlation_matrix_weather = correlation_features_weather.corr()

# Correlation heatmap for Plant 1 Weather
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_weather.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Plant 1 Weather')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 5))


axs[0].plot(p1_gen['DATE_TIME'], p1_gen['DC_POWER'])
axs[0].set_xlabel('Date')
axs[0].set_ylabel('kW')
axs[0].set_title('DC Power - Plant 2')
axs[0].tick_params(axis='x', rotation=45)


axs[1].plot(p1_gen['DATE_TIME'], p1_gen['AC_POWER'], color='#2b2924')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('kW')
axs[1].set_title('AC Power - Plant 2')
axs[1].tick_params(axis='x', rotation=45)


plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 5))
plt.plot(p1_gen['DATE_TIME'], p1_gen['DC_POWER'], color='#cc7e08', label='DC Power')
plt.plot(p1_gen['DATE_TIME'], p1_gen['AC_POWER'], color='#2b2924', label='AC Power')
plt.xlabel('Date')
plt.ylabel('kW')
plt.title('Power Output - Plant 1')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
